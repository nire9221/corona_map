```
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import re

url = 'http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=11&ncvContSeq=&contSeq=&board_id=&gubun='
table = pd.read_html(url,encoding="UTF-8")
len(table)

confirmed_case_age = table[5]
confirmed_case_age_range=table[5]['구분']
confirmed_case_age_count=table[5]['확진자(%)']

temp=[]
regex=re.compile(r'.+\(')
for data in confirmed_case_age_count:
    a=re.findall(regex,data)
    for num in a:
        b=int(num[:-2].replace(',',''))
        temp.append(b)
    confirmed_case_age_count=temp
        
# confirmed_case_age_range
# confirmed_case_age_count

font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
fontprop=fm.FontProperties(fname=font_path,size=10)
# plt.rc('font',family='NanumGothic')
# plt.rcParams['font.family']='NanumGothic'
# %config InlineBackend.figure_format = 'retina'
plt.rc('font', family=fontprop.get_name())
fm._rebuild()

plt.rcParams['axes.unicode_minus'] = False  

plt.figure(figsize=(10,10))
data = confirmed_case_age_count
categories = confirmed_case_age_range
plt.title('age-confirmed case')
plt.pie(data, labels=categories,autopct='%.1f%%',shadow=True,textprops={'fontsize':18})
plt.legend()
plt.show()



```
